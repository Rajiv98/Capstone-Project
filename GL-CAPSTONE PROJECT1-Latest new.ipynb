```python
#fills missing values in the CREDIT_SCORE column by using the mean CREDIT_SCORE of groups defined by the combination of EDUCATION and INCOME. 
df['CREDIT_SCORE']=df['CREDIT_SCORE'].fillna(df.groupby(['EDUCATION','INCOME'])['CREDIT_SCORE'].transform('mean'))
```

# OUTLIER TREATMENT

```python
df.select_dtypes(np.number)
```

```python
df.select_dtypes(include='object')
```

```python
for i in list(df.select_dtypes(np.number).columns):
    sns.boxplot(df[i])
    plt.show()
```

```python
# Percentage of Outlier Present in each Column

for i in list(df.select_dtypes(np.number).columns):
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr= q3-q1
    ul = q3+(1.5*iqr)
    ll = q1-(1.5*iqr)
    outliers = df[(df[i]>ul) | (df[i]<ll)]
    outliers_per = round(len(outliers) / df.shape[0],2)
    print(f'{i}: {outliers_per:.2%}')
```

```python
#This process limits the influence of outliers by capping them at the calculated limits.
for i in list(df.select_dtypes(np.number).columns):
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    ul=q3+(1.5*iqr)
    ll=q1-(1.5*iqr)
    df[i]=df[i].apply(lambda x:ll if x<ll else ul if x>ul else x )
```
